```python
import pandas as pd
```

```python
data_al_1_0 = pd.read_csv('Almaty_1_0.csv')
data_al_2_1 = pd.read_csv('Almaty_2_1.csv')
data_al_2_0 = pd.read_csv('Almaty_2_0.csv')
data_al_2_2 = pd.read_csv('Almaty_2_2.csv')
```

```python
Almaty = pd.concat([data_al_1_0, data_al_2_1, data_al_2_0, data_al_2_2], ignore_index=True)
```

```python
data_as_1_0 = pd.read_csv('Astana_1_0.csv')
data_as_2_1 = pd.read_csv('Astana_2_1.csv')
data_as_2_0 = pd.read_csv('Astana_2_0.csv')
data_as_2_2 = pd.read_csv('Astana_2_2.csv')
```

```python
Astana = pd.concat([data_as_1_0, data_as_2_1, data_as_2_0, data_as_2_2], ignore_index=True)
```

```python
data_sh_1_0 = pd.read_csv('Shymkent_1_0.csv')
data_sh_2_1 = pd.read_csv('Shymkent_2_1.csv')
data_sh_2_0 = pd.read_csv('Shymkent_2_0.csv')
data_sh_2_2 = pd.read_csv('Shymkent_2_2.csv')
```

```python
data_ak_1_0 = pd.read_csv('Aktau_1_0.csv')
data_ak_2_1 = pd.read_csv('Aktau_2_1.csv')
data_ak_2_0 = pd.read_csv('Aktau_2_0.csv')
data_ak_2_2 = pd.read_csv('Aktau_2_2.csv')
```
